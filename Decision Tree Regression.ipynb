```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
ds.info()
```

```python
#dividing depend and independ:
x=ds[['Level']]
y=ds[['Salary']]
```

```python

#algorithm of decision tree regression (algorithm fit)
from sklearn.tree import DecisionTreeRegressor,plot_tree
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
myinput=pd.DataFrame(data=[[6]],columns=['Level'])
reg.predict(myinput)
```

```python
plt.figure(figsize=(20,5))
plot_tree(reg,feature_names=['Feature 1'],filled=True)
plt.title("Decision Tree Visualiztion")
plt.show()
```
