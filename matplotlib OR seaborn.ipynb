```python
import matplotlib.pyplot as plt
x=['a','b','c','d','e']
y=[7,35,56,12,78]
plt.figure(figsize=(10,5))
plt.plot (x,y,color='green',marker='o')
plt.title("line chart")
plt.xlabel("name of people")
plt.ylabel("age of people")
for i in range(len(y)):
    plt.text(x[i],y[i] + 1,y[i],fontsize=12,ha='center')
plt.show()    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data=pd.DataFrame({'category':['a','b','c','a','a','b']})
plt.figure(figsize=(7,5))
sns.set_style("whitegrid")
sns.countplot(x='category',data=data ,palette='pastel')

plt.title("count plot",fontsize=14,fontweight='bold')
plt.xlabel("category")
plt.ylabel("count")
plt.show()

```

```python
#10)bar plot:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Values': [10, 10, 15]
})

# Set style and color palette
sns.set_style("ticks")  # Options: white, dark, whitegrid, darkgrid, ticks
sns.set_palette("colorblind")   # Try: deep, muted, bright, pastel, dark, colorblind

# Create the plot
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Category', y='Values', data=data)

# Add value labels on top of bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height()}', 
                     (p.get_x() + p.get_width() / 2, p.get_height()), 
                     ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Customize title and labels
plt.title("Bar Plot of Categories", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Optional: Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Remove top and right borders
sns.despine()

#count plot-A count plot is a type of bar plot that show the count of obs cat
# sns.countplot(x='category',data = data)
# plt.title("count plot")
# plt.show()
# Show the plot
plt.show()
```

```python
#14)strip plot:
import seaborn as sns 
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
sns.stripplot(x='day',y='total_bill',data=df)
plt.title("strip plot of total bill by day")
plt.show()
```

```python
#15)rug plot:
import seaborn as sns
import matplotlib.pyplot as plt


#load dataset:
df=sns.load_dataset("tips")

#create rug
sns.rugplot(x='total_bill',data=df)

plt.title("rug plot of total bill")
plt.show()
```

```python
#16)join plot:
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("tips")
sns.jointplot(x='total_bill',y='tips',data=df)
plt.title("join plot")
plt.show()
```
