```python
#credit and movies merged using their ids
merge_df = pd.merge(movies, credits, left_on = 'id' , right_on = 'movie_id', how = 'inner')
merge_df.shape
```

```python
#imp columns separated from the df and made new_df
new_df = merge_df[['id' , 'genres', 'keywords' , 'overview' , 'production_companies' , 'title_x', 'cast' , 'crew']]
```

```python
new_df
```

```python
#renamed column name as per convenience
new_df = new_df.rename(columns = {'title_x':'title'})
```

```python
(new_df.loc[0, 'keywords'])
```

```python
#abstract syntax tree
import  ast
x = ast.literal_eval(new_df.loc[0, 'keywords'])
x
```

```python

#by making function define
import ast 
def get_names(obj):
    x = ast.literal_eval(obj)
    keywords =[]
    for r in x:
        name = r['name'].replace(" ", "_").lower().strip() #to make tags, we added '_' to each string so it becomes a single string as tag
        if name not in keywords:
            keywords.append(name)
    return ",".join(keywords)   
```

```python
new_df['keywords'] = new_df['keywords'].apply(get_names)
```

```python
new_df['genres'] = new_df['genres'].apply(get_names)
```

```python
new_df['cast'] = new_df['cast'].apply(get_names)
```

```python
new_df['crew'] = new_df['crew'].apply(get_names)
```

```python
new_df['production_companies'] = new_df['production_companies'].apply(get_names)
```

```python
new_df.head(1)
```

```python
#lambda-apply given operation on each record in respective column
new_df['overview'] = new_df['overview'].apply(lambda x:str(x).replace(",","").replace(" ",",").lower())
new_df['overview']
```

```python
new_df['tags'] = new_df['genres'] + "," +  new_df['keywords']  + "," +  new_df['overview'] + "," +  new_df['cast'] + "," +  new_df['crew'] + "," + new_df['production_companies']
new_df
```

```python
final_df = new_df[["id", "title", "tags"]]
final_df
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.drop_duplicates(inplace = True)
```

```python
#remove stopwords natural language toolkit
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#download stopwords if not alredy exists
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def remove_stop_words(text):
    #get the english stopword list
    stop_words = set(stopwords.words('english'))
    #tokenize the sentence
    words = text.split(',')
    #remove stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]
    #convert back to string
    filtered_text = ','.join(filtered_words)
    return filtered_text
```

```python
final_df['tags'] = final_df['tags'].apply(remove_stop_words)
```
