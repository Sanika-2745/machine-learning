```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
# import warnings
# warnings.filterwarnings('ignore')

```

```python
ds=pd.read_csv('startup_data.csv')
ds
```

```python
ds.info()
```

```python
#diving into dependent and independent:
x=ds[['R&D Spend','Administration','Marketing Spend','State']]
y=ds[['Profit']]
```

```python
#encoded state column using onehotencoder:
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded_data=ohe.fit_transform(x[['State']])
encoded_data
```

```python
# import matplotlib.pyplot as plt
```

```python
column_name=ohe.get_feature_names_out(['State'])
encoded_df=pd.DataFrame(encoded_data,columns=column_name)
encoded_df
```

```python
#joined x(independed data variable)and encoded dataframe:
x=pd.concat([x,encoded_df],axis=1)
x=x.drop(columns=['State'])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=5,random_state=0)
x_test
```

```python
#let train model:
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
data=[[165349.20, 136897.80, 471784.10, 0, 0, 1]]
columns=x.columns
myinput=pd.DataFrame(column=columns,data=data)
result=reg.predict(myinput)
result
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
